```python
import pandas as pd
import gradio as gr

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
tracks_clean = pd.read_csv('./data/tracks_clean.csv')
```

```python
tracks_clean
```

```python
X = tracks_clean.drop(columns=['popularity', 'duration_ms', 'release_date', 'key', 'mode', 'speechiness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'popularity_category'], axis=1)
y = tracks_clean['popularity_category']

#test/train/split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.23,
                                                    random_state=42,
                                                    stratify=y
                                                   )
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
def predict_popularity(explicit, danceability, energy, loudness, acousticness):
    tracks = pd.DataFrame.from_dict({'explicit':[explicit],
                                     'danceability':[danceability],
                                     'energy':[energy],
                                     'loudness':[loudness],
                                     'acousticness':[acousticness]})
    
    predict = rf.predict_proba(tracks)[0]
    
    return {rf.classes_[i]: predict[i] for i in range(3)}
```

```python
explicit = gr.components.Slider(minimum=0, maximum=1, label='explicit')

danceability = gr.components.Slider(minimum=0, maximum=1, label='danceability')

energy = gr.components.Slider(minimum=0, maximum=1, label='energy')

loudness = gr.components.Slider(minimum=-60, maximum=5.376, label='loudness')

acousticness = gr.components.Slider(minimum=0, maximum=1, label='acousticness')

gr.Interface(predict_popularity, [explicit, danceability, energy, loudness, acousticness],
            'label', live=True, interpretation='default').launch(debug=True)

#https://towardsdatascience.com/understanding-feature-importance-and-how-to-implement-it-in-python-ff0287b20285
```
